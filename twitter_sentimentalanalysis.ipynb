# Importing libraries

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
```

# Extracting Tweets using snscrape

```python
query = "bigbilliondays"
l = []
length = 1000

for i in sntwitter.TwitterSearchScraper(query).get_items():
    if len(l) == length:
        break
    l.append(i)
l
```

```python
df = pd.DataFrame(l)
```

```python
df
```

```python
import re
```

```python
l1 = list(df["content"])
l1
```

###### Removing emojis,symbols,other language text using RE

```python

lines = [re.sub(r'[^A-Za-z0-9]+', ' ', x) for x in l1]

lines

lines2 = []

for word in lines:
    if word != '':
        lines2.append(word)
```

```python
lines2
```

```python
df2 = pd.DataFrame(lines2)
```

```python
df2
```

```python
df['content'] = df2
df
```

```python
df.head(10)
```

# Importing NLTK

```python
# Import stopwords
import nltk
from nltk.corpus import stopwords

# Import textblob
from textblob import Word, TextBlob
```

```python
# Calculate polarity
df['polarity'] = df['content'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity'] = df['content'].apply(lambda x: TextBlob(x).sentiment[1])
```

```python
df
```

###### Classifying Sentiment based on polarity values

```python
df["sentiment"] = np.where(df['polarity'] >= 0 , 'Positive', 'Negative')
df.head()
```

```python
new_df = df[['date','content','hashtags','polarity','subjectivity','sentiment']].copy()
```
